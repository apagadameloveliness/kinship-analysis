# Installing and setting up libraries and environment

```r
install.packages("RColorBrewer"); library(RColorBrewer);
install.packages("scales"); library(scales);
palette(brewer.pal(4, "Dark2"))
```

# Extracting the simulation LR data

```r
data <- read.table("data/Duo_94iisnp_SimLRs.txt", skip = 5)
head(data)
```

```r
hist(log10(data$V1), col=alpha(1, 0.8), prob=T, xlab="Log10 LR", main="", xlim=c(min(log10(data$V2)),max(log10(data$V1))), ylim=c(0,0.8) )

hist(log10(data$V2),add=T, col=alpha(2, 0.8), prob=T)

points(density(log10(data$V1)), col=alpha(1, 0.8), lty=2, lwd=2, type='l')

points(density(log10(data$V2)), col=alpha(2, 0.8), lty=2, lwd=2, type='l')

abline(v=mean(log10(data$V1)), lwd=2, lty=2, col="black")

abline(v=mean(log10(data$V2)), lwd=2, lty=2, col="black")

par(xpd=T)

text(x=mean(log10(data$V1))+1, y=0.8, labels=c("H1-Duo"), col=alpha(1,0.8), cex=1.2)

text(x=mean(log10(data$V2))-1, y=0.8, labels=c("H2-Unrelated"), col=alpha(2,0.8), cex=1.2)

par(xpd=F)

dev.new()
```
