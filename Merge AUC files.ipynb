# Merge all AUC summary tables

```r
# Set your main folder here
main_dir <- "../8_ROC"

# Pattern match and read
files <- list.files(
  path = main_dir,
  pattern = "^AUC_summary.*\\.txt$",
  recursive = TRUE,
  full.names = TRUE
)

if (length(files) == 0) stop("No matching AUC_summary*.txt files found")

# Adjust `sep`/`header` as needed for your files
merged <- do.call(
  rbind,
  lapply(files, function(f) {
    df <- read.table(f, sep = "\t", header = TRUE, stringsAsFactors = FALSE, check.names = FALSE)
    df$source_file <- strsplit(basename(f), "_")[[1]][3]
    df
  })
)

# Save merged table
write.table(merged, file = file.path(main_dir, "AUC_all_merged.txt"),
            sep = "\t", quote = FALSE, row.names = FALSE)

cat("Merged", length(files), "files into", file.path(main_dir, "AUC_summary_merged.txt"), "\n")
```
